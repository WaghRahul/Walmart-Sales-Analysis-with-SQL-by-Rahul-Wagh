#***Dataset Description:***
##***WalmartData***

The dataset consists of the following columns:

* `Invoice_ID`: Unique identifier for each invoice.
* `Branch`: Branch of the store where the sale took place.
* `City`: City where the store branch is located.
* `Customer_type`: Type of customer (e.g., Member, Normal).
* `Gender`: Gender of the customer.
* `Product_line`: Category of the product sold.
* `Unit_price`: Price per unit of the product.
* `Quantity`: Quantity of the product sold.
* `Tax_5%`: Tax applied on the sale.
* `Total`: Total amount of the sale.
* `Date`: Date of the transaction.
* `Time`: Time of the transaction.
* `Payment`: Payment method used (e.g., Cash, Credit Card).
* `cogs`: Cost of goods sold.
* `gross_margin_percentage`: Gross margin percentage.
* `gross_income`: Gross income from the sale.
* `Rating`: Customer rating for the purchase

##***Feature Engineering***

```python
%%sql
--Update the 'Date' column to match the desired format
UPDATE WalmartData SET Date = substr(Date, 7, 4) || '-' || substr(Date, 4, 2) || '-' || substr(Date, 1, 2);

--Query to verify the date format conversion
SELECT Date FROM WalmartData LIMIT 10;
```

```python
%%sql
--Add a new column to store the time of the day
select time, (case when `time` between "00:00:00" and "12:00:00" then "Morning"
                   when `time` between "12:01:00" and "16:00:00" then "Afternoon"
                   else "Evening"
              end) as Time_of_the_day
from WalmartData;

--Update the new column with the time of the day based on the time column
alter table WalmartData add column Time_of_the_day;
update WalmartData set Time_of_the_day = (case when `time` between "00:00:00" and "12:00:00" then "Morning"
                                           when `time` between "12:01:00" and "16:00:00" then "Afternoon"
                                           else "Evening"
                                       end);

--Verify the update by selecting the Time and Time_of_the_day columns
select Time, Time_of_the_day from WalmartData limit 10;
```

```python
%%sql
-- Add a new column to store the day name
ALTER TABLE WalmartData ADD COLUMN Day TEXT;

-- Update the new column with the day name based on the date
UPDATE WalmartData
SET Day = CASE
    WHEN strftime('%w', Date) = '0' THEN 'Sunday'
    WHEN strftime('%w', Date) = '1' THEN 'Monday'
    WHEN strftime('%w', Date) = '2' THEN 'Tuesday'
    WHEN strftime('%w', Date) = '3' THEN 'Wednesday'
    WHEN strftime('%w', Date) = '4' THEN 'Thursday'
    WHEN strftime('%w', Date) = '5' THEN 'Friday'
    WHEN strftime('%w', Date) = '6' THEN 'Saturday'
END;

-- Verify the update by selecting the Date and Day columns
SELECT Date, Day FROM WalmartData LIMIT 10;
```

```python
%%sql
-- Add a new column to store the month name
ALTER TABLE WalmartData ADD COLUMN Month TEXT;

-- Update the new column with the month name based on the date
UPDATE WalmartData
SET Month = CASE
    WHEN strftime('%m', Date) = '01' THEN 'January'
    WHEN strftime('%m', Date) = '02' THEN 'February'
    WHEN strftime('%m', Date) = '03' THEN 'March'
    WHEN strftime('%m', Date) = '04' THEN 'April'
    WHEN strftime('%m', Date) = '05' THEN 'May'
    WHEN strftime('%m', Date) = '06' THEN 'June'
    WHEN strftime('%m', Date) = '07' THEN 'July'
    WHEN strftime('%m', Date) = '08' THEN 'August'
    WHEN strftime('%m', Date) = '09' THEN 'September'
    WHEN strftime('%m', Date) = '10' THEN 'October'
    WHEN strftime('%m', Date) = '11' THEN 'November'
    WHEN strftime('%m', Date) = '12' THEN 'December'
END;

-- Verify the update by selecting the Date and Month columns
select date, Month from WalmartData limit 10;
```

##***Exploratory Data Analysis (EDA)***
- Exploratory data analysis is done to answer the listed questions

##***1.Retrieve all columns for sales made in a specific branch (e.g., Branch 'A').***

```python
%%sql
--Query to retrieve all columns for sales made in a specific branch (e.g., Branch 'A')
SELECT * FROM WalmartData WHERE Branch='A';
```

##***2. Find the total sales for each product line.***

```python
%%sql
--Query to calculate the total sales for each product line
SELECT Product_line, SUM(Total) AS Total_Sales FROM WalmartData GROUP BY Product_line;
```

##***3. List all sales transactions where the payment method was 'Cash'.***

```python
%%sql
--Query to list all sales transactions where the payment method was 'Cash'
SELECT * FROM WalmartData WHERE Payment='Cash';
```

##***4. Calculate the total gross income generated in each city.***

```python
%%sql
--Query to calculate the total gross income generated in each city
SELECT City, SUM(gross_income) AS Total_Gross_Income FROM WalmartData GROUP BY City;
```

##***5. Find the average rating given by customers in each branch.***

```python
%%sql
--Query to find the average rating given by customers in each branch
SELECT Branch, AVG(Rating) AS Average_Rating FROM WalmartData GROUP BY Branch;
```

##***6. Determine the total quantity of each product line sold.***

```python
%%sql
--Query to determine the total quantity of each product line sold
SELECT Product_line, SUM(Quantity) AS Total_Quantity FROM WalmartData GROUP BY Product_line;
```

##***7. List the top 5 products by unit price.***

```python
%%sql
--Query to list the top 5 products by unit price
SELECT Product_line, Unit_price FROM WalmartData ORDER BY Unit_price DESC LIMIT 5;
```

##***8. Find sales transactions with a gross margin percentage greater than 30%.***

```python
%%sql
--Query to find sales transactions with a gross margin percentage greater than 30%
SELECT * FROM WalmartData WHERE gross_margin_percentage>3.0;
```

##***9. Retrieve sales transactions that occurred on weekends.***

```python
%%sql
--Query to retrieve sales transactions that occurred on weekends
SELECT * FROM WalmartData WHERE Day IN ('Sunday', 'Saturday');
```

##***10. Calculate the total sales and gross income for each month.***

```python
%%sql
--Query to calculate the total sales and gross income for each month
SELECT Month, SUM(Total) AS Total_Sales, SUM(gross_income) AS Total_Gross_Income FROM WalmartData GROUP BY Month;
```

##***11. Find the number of sales transactions that occurred after 6 PM.***

```python
%%sql
--Query to find the number of sales transactions that occurred after 6 PM
SELECT COUNT(*) AS Number_of_Transactions FROM WalmartData WHERE strftime('%H:%M:%S', Time) > '18:00:00';
```

##***12. List the sales transactions that have a higher total than the average total of all transactions.***

```python
%%sql
--Query to list the sales transactions that have a higher total than the average total of all transactions
SELECT * FROM WalmartData WHERE Total > (SELECT AVG(Total) FROM WalmartData);
```

##***13. Calculate the cumulative gross income for each branch by date. 15. Find the total cogs for each customer type in each city.***

```python
%%sql
--Query to calculate the cumulative gross income for each branch by date
SELECT Branch, Date, SUM(cogs) AS Total_cogs FROM WalmartData GROUP BY Branch, Date;
```
